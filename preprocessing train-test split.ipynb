```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import time

# Muat dataset
start_time = time.time()
data = pd.read_csv('diabetes.csv')
print(f"Loading dataset: {time.time() - start_time:.4f} seconds")

# Pisahkan fitur dan label
start_time = time.time()
X = data.drop('Outcome', axis=1)
y = data['Outcome']
print(f"Separating features and labels: {time.time() - start_time:.4f} seconds")

# Bagi dataset menjadi data latih dan data uji dengan rasio 70:30 dan 80:20
start_time = time.time()
X_train_70, X_test_30, y_train_70, y_test_30 = train_test_split(X, y, test_size=0.30, random_state=42)
print(f"Splitting dataset 70:30: {time.time() - start_time:.4f} seconds")

start_time = time.time()
X_train_80, X_test_20, y_train_80, y_test_20 = train_test_split(X, y, test_size=0.20, random_state=42)
print(f"Splitting dataset 80:20: {time.time() - start_time:.4f} seconds")

# Normalisasi fitur
start_time = time.time()
scaler = StandardScaler()
X_train_70 = scaler.fit_transform(X_train_70)
X_test_30 = scaler.transform(X_test_30)
print(f"Normalizing features for 70:30 split: {time.time() - start_time:.4f} seconds")

start_time = time.time()
X_train_80 = scaler.fit_transform(X_train_80)
X_test_20 = scaler.transform(X_test_20)
print(f"Normalizing features for 80:20 split: {time.time() - start_time:.4f} seconds")
```
