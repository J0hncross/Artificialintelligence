```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Load dataset with specified encoding
data = pd.read_csv('./Dataset/SeoulBikeData.csv', encoding='ISO-8859-1')

# Select relevant features for the regression (dropping categorical and date-related features for now)
features = ['Hour', 'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)',
            'Visibility (10m)', 'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)',
            'Rainfall(mm)', 'Snowfall (cm)']

X = data[features]
y = data['Rented Bike Count']

# Split the dataset into training and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the SVR model with a linear kernel
model = SVR(kernel='linear')
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate the coefficient of determination (R-squared)
r2 = r2_score(y_test, y_pred)

# Calculate the mean absolute error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Get the model coefficients and intercept
coefficients = model.coef_ if hasattr(model, 'coef_') else None
intercept = model.intercept_

# Display the results
print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R-squared (Coefficient of Determination): {r2}')
print(f'Model Coefficients: {coefficients}')
print(f'Model Intercept: {intercept}')

# Display the kernel parameter
print(f'Kernel used: {model.kernel}')
```
